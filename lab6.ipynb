```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is in a DataFrame named df
df = pd.read_sql(sql,conn)

# Calculate the year over year income growth rate for each state
df = df.sort_values(['name', 'year'])
df['income_growth'] = df.groupby('name')['income'].pct_change() * 100

# Pivot the data for visualization
pivot_df = df.pivot(index='year', columns='name', values='income_growth')

# Plot the data
plt.figure(figsize=(10,8))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='', linewidth=1, label=column)

plt.legend(loc='upper right', ncol=2)
plt.title('Year-Over-Year Income Growth Rate by State')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate (%)')
plt.grid(True)
plt.show()

# This code first sorts the DataFrame by `name` (state) and `year`, then calculates the year-over-year income growth rate for each state. It then pivots the DataFrame to make the `year` column the index and each state a separate column, which is an appropriate format for line chart visualization. The chart is then plotted with the year on the x-axis and the income growth rate on the y-axis. Each state is represented by a separate line on the chart. The chart includes a legend in the upper right corner, a title, labels for the x- and y-axes, and a grid for easier reading of the data points. Note: replace 'your_data_file.csv' with the actual path to your csv file.
```

```python
import matplotlib.pyplot as plt

# Sort DataFrame by 'name' and 'year'
df.sort_values(['name', 'year'], inplace=True)

# Calculate Year over Year Growth for each state
df['yoy_growth'] = df.groupby('name')['income'].pct_change()

# Pivot the DataFrame to prepare for line chart
pivot_df = df.pivot(index='year', columns='name', values='yoy_growth')

# Plotting
plt.figure(figsize=(12,8))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Year over Year Income Growth Rate by State", loc='left', fontsize=12, fontweight=0)
plt.xlabel("Year")
plt.ylabel("Growth Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by state and year to calculate the year over year income growth rate
df_grouped = df.groupby(["name", "year"])["income_growth"].mean().reset_index()

# Pivot the data to get years as columns and states as rows
df_pivot = df_grouped.pivot(index="name", columns="year", values="income_growth")

# Plotting
plt.figure(figsize=(15,10))
for state in df_pivot.index:
    plt.plot(df_pivot.columns, df_pivot.loc[state], label=state)

plt.title('Year Over Year Income Growth Rate per State')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate (%)')
plt.legend(loc='best')
plt.grid()
plt.show()
```
